```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('EdStatsData.csv')
```

```python
data.columns
```

```python
# Displaying the list of regions
```

```python
data["Country Name"].unique()
```

```python
# Filtering to contain only the data about the Europe and Central Asia Region
```

```python
dataECS = data[data["Country Code"].isin(["ECS","WLD"])]
```

```python
dataECS
```

```python
# Creating a set for the years we will work on
```

```python
cols = dataECS.columns[dataECS.columns.slice_indexer('2000','2015')]
```

```python
cols
```

```python
# Filtering so we remain only with rows that contains data for the columns in set
```

```python
dataECS = dataECS.dropna(subset=cols,how='all')
```

```python
dataECS = dataECS.drop(dataECS.loc[:, '2015':'2100'].columns, axis = 1)
```

```python
dataECS = dataECS.drop(dataECS.loc[:, '1970':'2000'].columns, axis = 1)
```

```python
dataECS = dataECS.drop('Unnamed: 69',axis=1)
```

```python
dataECS.set_index('Indicator Code',inplace=True)
```

```python
dataECS
```

```python
# Creating a csv with the filtered data
```

```python
dataECS.to_csv('ECS.csv')
```

```python
# Filtering data for a csv where the indicators are applied for  Europe and Central Asia separetly
# Steps below are identical with the one above
```

```python
countryData = pd.read_csv('EdStatsCountry.csv')
```

```python
countryDataE = countryData[countryData['Region'] == 'Europe & Central Asia']
countries = countryDataE['Table Name'].to_list()
dataCountries = data[data["Country Name"].isin(countries)]
```

```python
dataCountries = dataCountries.dropna(subset= cols,how='all')
dataCountries = dataCountries.drop('Unnamed: 69',axis=1)
dataCountries = dataCountries.drop(dataCountries.loc[:, '1970':'2000'].columns, axis = 1)
dataCountries = dataCountries.drop(dataCountries.loc[:, '2015':'2100'].columns, axis = 1)
```

```python
dataCountries["region"] = "Europe and Central Asia"
```

```python
dataCountries.to_csv('ECSCountries.csv')
```

```python
#CountriesForAllRegions
```

```python
ecs = dataCountries
```

```python
ecs
```

```python
countryDataA = countryData[countryData['Region'] == 'East Asia & Pacific']
countries = countryDataA['Table Name'].to_list()
dataCountries = data[data["Country Name"].isin(countries)]
```

```python
dataCountries = dataCountries.dropna(subset= cols,how='all')
dataCountries = dataCountries.drop('Unnamed: 69',axis=1)
dataCountries = dataCountries.drop(dataCountries.loc[:, '1970':'2000'].columns, axis = 1)
dataCountries = dataCountries.drop(dataCountries.loc[:, '2015':'2100'].columns, axis = 1)
```

```python
dataCountries["region"] = "East Asia & Pacific"
```

```python
eas = dataCountries
```

```python
countryDataN = countryData[countryData['Region'] == 'North America']
countries = countryDataN['Table Name'].to_list()
dataCountries = data[data["Country Name"].isin(countries)]
```

```python
dataCountries = dataCountries.dropna(subset= cols,how='all')
dataCountries = dataCountries.drop('Unnamed: 69',axis=1)
dataCountries = dataCountries.drop(dataCountries.loc[:, '1970':'2000'].columns, axis = 1)
dataCountries = dataCountries.drop(dataCountries.loc[:, '2015':'2100'].columns, axis = 1)
```

```python
dataCountries["region"] = "North America"
```

```python
nac = dataCountries
```

```python
frames = [ecs, eas, nac]
```

```python
result = pd.concat(frames)
```

```python
result.to_csv('RegionCountries.csv')
```
